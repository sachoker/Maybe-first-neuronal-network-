```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans, DBSCAN
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12, 5)
```

```python
x = pd.read_csv("данные.csv", sep = ',', encoding = "windows-1251")
f = x.values

a = KMeans(n_clusters = 9).fit(x)
labels = a.labels_
centroids = a.cluster_centers_
print(a.cluster_centers_)
print(a.labels_)
```

```python
a = np.meshgrid(a)
```

```python
colors = ["g.","r.","c.","y.","b.", "k.", "gold", "purple", "magenta"]

for i in range(len(f)):
    print("coordinate:",f[i], "label:", labels[i])
    plt.plot(f[i][0], f[i][1], colors[labels[i]], markersize = 10)
#https://habr.com/ru/post/304214/

plt.scatter(centroids[:, 0],centroids[:, 1], marker = "x", s=150, linewidths = 5, zorder = 10)

plt.show()
```
