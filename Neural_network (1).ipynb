```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans, DBSCAN
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12, 5)
```

```python
def filter(x):
    y = x.copy()
    for i in x:
        cnt = 0
        for j in x[i]:
                if j:
                    break
                else:
                    cnt += 1
                if cnt == len(x[i]) or cnt == len(x[i]) - 1:
                    print(i)
                    y = y.drop(labels = i, axis = 1)
                    break
    return y
```

```python
x = pd.read_csv("данные.csv", sep = ',', encoding = "windows-1251")
```

```python
pca = PCA(n_components = 3)
x = filter(x)
X = pca.fit_transform(x)
f = x.values
```

```python
a = KMeans(n_clusters = 9).fit(X)
labels = a.labels_
centroids = a.cluster_centers_
```

```python
colors = ["g.","r.","c.","y.","b.", "k.", "gold", "purple", "magenta"]

fig = plt.figure()
ax = fig.gca(projection='3d')

print(type(X[1][0]))
for i in range(len(f)):
    print("coordinate:",X[i], "label:", labels[i])
    ax.scatter(X[i][2], X[i][1], X[i][2], colors[labels[i]])
#ax.scatter(list(centroids[:, 0]),list(centroids[:, 1]), list(centroids[:, 2]),  marker = "x", linewidths = 5, zorder = 30)



plt.show()
```
